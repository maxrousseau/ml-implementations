```python
# draw dots with PIL
def draw_dots(image, dot_list, col):
    
    for i in dot_list:
        i = i.flatten().astype(int)
        x0 = i[0] - 1
        y0 = i[1] + 1
        x1 = i[0] + 1
        y1 = i[1] - 1
        newim = ImageDraw.Draw(image)
        newim.rectangle((x0,y0,x1,y1), fill=col)
    
draw_dots(test_im, ldmk, col='red')
plt.imshow(test_im)
```
