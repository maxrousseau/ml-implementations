```python
# testing hacked MTCNN
# original code: https://github.com/timesler/facenet-pytorch
from mtcnn import MTCNN
from PIL import Image, ImageDraw

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
# load and view image
test_im = Image.open('./data/photo.jpg')
test_im = test_im.rotate(-90)
test_im = test_im.resize((test_im.width//6, test_im.height//6))
test_npim = np.array(test_im)
plt.imshow(test_im)
```

```python
mtcnn_md = MTCNN()
box, prob = mtcnn_md.detect(test_npim)
box = box[0]
print(box)
```

```python
# don't forget to incorporate this into the source code
def drawbox_pil(im, box, col):
    """draw box with pillow
    """
    x0 = box[0] 
    y0 = box[1]
    x1 = box[2]
    y1 = box[3]
    newim = ImageDraw.Draw(im)
    newim.rectangle((x0,y0,x1,y1), outline=col)


drawbox_pil(test_im, box, 'blue')
plt.imshow(test_im)
```
